### write a program in python to draw different geometric shapes and to write text on images using OpenCV

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img=np.zeros((500,500,3),dtype='uint8')
```

```python
plt.imshow(img)
plt.imshow(img)
plt.show()
```

```python
cv2.line(img,(30,30),(300,200),(0,0,255),3)
cv2.rectangle(img,(30,30),(300,200),(0,255,0),3)
cv2.circle(img,(100,300),40,(100,42,80),4)
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,'computer vision',(10,350),
           font,0.8,(0,255,0),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
